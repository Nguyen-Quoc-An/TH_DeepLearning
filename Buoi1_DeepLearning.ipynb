```python
# Dùng tf.function bọc hàm thường
def a_regular_function(x, y, b):
    x = tf.matmul(x, y)
    x = x + b
    return x
a_function_that_uses_a_graph = tf.function(a_regular_function)
x1 = tf.constant([[1.0,2.0]])
y1 = tf.constant([[2.0],[3.0]])
b1 = tf.constant(4.0)
orig_value = a_regular_function(x1, y1, b1).numpy()
tf_value   = a_function_that_uses_a_graph(x1, y1, b1).numpy()
print(orig_value); print(tf_value)
# Dùng decorator @tf.function
def inner_function(x, y, b):
    x = tf.matmul(x, y)
    x = x + b
    return x
@tf.function
def outer_function(x):
    y = tf.constant([[2.0],[3.0]])
    b = tf.constant(4.0)
    return inner_function(x, y, b)
rs = outer_function(tf.constant([[1.0,2.0]])).numpy()
print(rs)
```

```python
# Tạo layer SimpleModule
class SimpleModule(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.a_variable = tf.Variable(5.0, name="train_me")
        self.non_trainable_variable = tf.Variable(5.0, trainable=False, name="do_not_train_me")
    def __call__(self, x):
        return self.a_variable * x + self.non_trainable_variable
simple_module = SimpleModule(name="simple")
rs = simple_module(tf.constant(5.0))
print(rs)
print('Train variables', simple_module.trainable_variables)
print('All variables', simple_module.variables)
# Model với nhiều layer Dense
class Dense(tf.Module):
    def __init__(self, in_features, out_features, name=None):
        super().__init__(name=name)
        self.w = tf.Variable(tf.random.normal([in_features, out_features]), name='w')
        self.b = tf.Variable(tf.zeros([out_features]), name='b')
    def __call__(self, x):
        y = tf.matmul(x, self.w) + self.b
        return tf.nn.relu(y)
class SequentialModule(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.dense_1 = Dense(in_features=3, out_features=3)
        self.dense_2 = Dense(in_features=3, out_features=2)
    def __call__(self, x):
        x = self.dense_1(x)
        return self.dense_2(x)
my_model = SequentialModule(name="the_model")
print("Model results:", my_model(tf.constant([[2.0, 2.0, 2.0]])))
```

```python
# Lưu checkpoint
chkp_path = "my_checkpoint"
checkpoint = tf.train.Checkpoint(model=my_model)
checkpoint.write(chkp_path)
# Phục hồi checkpoint
new_model = SequentialModule()
new_checkpoint = tf.train.Checkpoint(model=new_model)
new_checkpoint.restore("my_checkpoint")
print(new_model(tf.constant([[2.0, 2.0, 2.0]])))
```

```python
#12.
x = tf.random.uniform([3,2,2], minval=1, maxval=10)
y = tf.random.uniform([3,2,2], minval=1, maxval=10)
@tf.function
def cong(a, b): return tf.add(a, b)
@tf.function
def tru(a, b): return tf.subtract(a, b)
@tf.function
def nhan(a, b): return tf.multiply(a, b)
@tf.function
def chia(a, b): return tf.divide(a, b)
print('x+y:', cong(x, y))
print('x-y:', tru(x, y))
print('x*y:', nhan(x, y))
print('x/y:', chia(x, y))
```

```python
#13.
class SumModel(tf.Module):
    def __call__(self, x, y):
        return tf.add(x, y)
model_sum = SumModel()
x = tf.constant([1.0, 2.0, 3.0])
y = tf.constant([4.0, 5.0, 6.0])
print('Tổng:', model_sum(x, y))
```

```python
#14.
class DiffProductModel(tf.Module):
    def __call__(self, x, y):
        hieu = tf.subtract(x, y)
        tich = tf.multiply(x, y)
        return hieu, tich
model_dp = DiffProductModel()
x = tf.constant([6.0, 8.0, 10.0])
y = tf.constant([2.0, 4.0, 5.0])
hieu, tich = model_dp(x, y)
print('Hiệu:', hieu)
print('Tích:', tich)
```

```python
#15.
class ChuViModel(tf.Module):
    def __call__(self, a, b):
        return tf.multiply(tf.add(a, b), 2.0)
model_cv = ChuViModel()
a = tf.constant(5.0)
b = tf.constant(3.0)
print('Chu vi P =', model_cv(a, b).numpy())
```

```python
#16.
class DienTichModel(tf.Module):
    def __call__(self, a, b, c):
        s_xq = tf.multiply(tf.multiply(tf.add(a, b), c), 2.0)
        s_tp = tf.add(tf.multiply(tf.multiply(a, b), 2.0), s_xq)
        return s_xq, s_tp

model_dt = DienTichModel()
a = tf.constant(4.0)
b = tf.constant(3.0)
c = tf.constant(5.0)
s_xq, s_tp = model_dt(a, b, c)
print('S xung quanh:', s_xq.numpy())
print('S toan phan:', s_tp.numpy())
```

```python
#17.
class SongSongModel(tf.Module):
    def __call__(self, AD, BC, AB, CD):
        dieu_kien_1 = tf.equal(AD, BC)
        dieu_kien_2 = tf.equal(AB, CD)
        ad_song_song_bc = tf.logical_and(dieu_kien_1, dieu_kien_2)
        return ad_song_song_bc
model_ss = SongSongModel()
AD = tf.constant(5.0); BC = tf.constant(5.0)
AB = tf.constant(4.0); CD = tf.constant(4.0)
ket_qua = model_ss(AD, BC, AB, CD)
print('AD // BC:', ket_qua.numpy())
```

```python
#18.
class HaiDayModel(tf.Module):
    def __call__(self, AD, BC, AB, CD):
        dk_ad_bang_bc = tf.equal(AD, BC)
        dk_ad_song_bc = tf.equal(AB, CD)
        ab_bang_cd = tf.logical_and(dk_ad_bang_bc, dk_ad_song_bc)
        return ab_bang_cd
model_hd = HaiDayModel()
AD = tf.constant(6.0); BC = tf.constant(6.0)
AB = tf.constant(3.0); CD = tf.constant(3.0)
ket_qua = model_hd(AD, BC, AB, CD)
print('AB = CD:', ket_qua.numpy())
```

```python
#19.
class TrungBinhModel(tf.Module):
    def __call__(self, AM, MB, AN, NC, BC):
        am_bang_mb = tf.equal(AM, MB)
        an_bang_nc = tf.equal(AN, NC)
        mn_la_trung_binh = tf.logical_and(am_bang_mb, an_bang_nc)
        MN = tf.divide(BC, 2.0)
        na_bang_nc = tf.logical_and(
            mn_la_trung_binh,
            tf.equal(AN, NC)
        )
        return na_bang_nc, MN
model_tb = TrungBinhModel()
AM = tf.constant(3.0); MB = tf.constant(3.0)
AN = tf.constant(4.0); NC = tf.constant(4.0)
BC = tf.constant(8.0)
na_nc, mn = model_tb(AM, MB, AN, NC, BC)
print('NA = NC:', na_nc.numpy())
print('MN =', mn.numpy(), '= BC/2 =', (BC/2).numpy())
```
